```python
import tkinter as tk
import math

class JumpingHeart:
    def __init__(self, root):
        self.root = root
        self.root.title("激动的心")  # 修改窗口标题
        self.root.geometry("400x400")
        
        # 创建画布
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()
        
        # 爱心的初始参数
        self.scale = 10  # 初始缩放比例
        self.grow = True  # 控制放大/缩小
        self.center_x, self.center_y = 200, 200  # 爱心中心坐标
        
        # 开始动画
        self.animate()

    def draw_heart(self, scale):
        """绘制爱心，scale控制爱心大小"""
        # 清空画布
        self.canvas.delete("all")
        
        # 笛卡尔心形公式：(x²+y²-1)³ - x²y³ = 0
        points = []
        for theta in range(0, 360, 2):
            angle = math.radians(theta)
            x = 16 * math.sin(angle)**3
            y = 13 * math.cos(angle) - 5 * math.cos(2*angle) - 2 * math.cos(3*angle) - math.cos(4*angle)
            
            # 缩放并平移到画布中心
            x = self.center_x + x * scale
            y = self.center_y - y * scale  # 负号是为了让爱心正过来
            points.append(x)
            points.append(y)
        
        # 绘制红色爱心：去掉生硬边框，用渐变思路画多层浅红轮廓实现柔和效果
        # 1. 最内层：纯红色填充（核心）
        self.canvas.create_polygon(points, fill="red", outline="", width=0)
        # 2. 外层1：浅一点的红色轮廓，宽度1，实现柔和过渡
        self.canvas.create_polygon(points, fill="", outline="#ff6666", width=1)
        # 3. 外层2：更浅的红色轮廓，宽度1，进一步柔化边缘
        self.canvas.create_polygon(points, fill="", outline="#ff9999", width=1)

    def animate(self):
        """实现跳动动画"""
        # 控制缩放比例，实现跳动效果
        if self.grow:
            self.scale += 0.2
            if self.scale >= 12:  # 最大尺寸
                self.grow = False
        else:
            self.scale -= 0.2
            if self.scale <= 8:  # 最小尺寸
                self.grow = True
        
        # 绘制当前大小的爱心
        self.draw_heart(self.scale)
        
        # 每隔20毫秒调用一次animate，形成动画
        self.root.after(20, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    app = JumpingHeart(root)
    root.mainloop()
```
